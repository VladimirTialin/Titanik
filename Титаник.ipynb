 В представленном файле имееютя следующе данные:
 1. **PassengerId** - идентификатор пассажира;
 2. **Survived** 	- выжившие;
 3. **Pclass** 	- класс пассажира;
 4. **Name** - Ф.И.О. пассажира;
 5. **Sex** - пол пассажира;
 6. **Age** - возраст пассажира;
 7. **SibSp** — число братьев, сестер или супругов на борту у человека;
 8. **Parch** — количество родителей, детей (в том числе приемных);
 9. **Ticket** - билет;
 10. **Fare** - плата за проезд;
 11. **Cabin** -  каюта;
 12. **Embarked** — порт посадки.

**Количество мужчин и женщин на борту**

```python
# Количество мужчин и женщин на борту
df.groupby('Sex')['PassengerId'].count()
```

**Возраст пассажиров:**

```python
print('Максимальный возраст: ',end='')
print(df['Age'].max())
print('Минимальный возраст: ',end='')
print(df['Age'].min())
print('Средний возраст: ',end='')
print(df['Age'].mean())
```

**Данные о самом старом пассжире**

```python
# Ф.И.О. самого старого пассажира
df[['Name', 'Age']].max()
```

**Данные о самом молодом пассажире**

```python
# Ф.И.О. самого молодого пассажира
df[['Name', 'Age']].min()
```

**Количество выживших мужчин и женщин**

```python
# Количество выживших пасажиров
df.groupby('Sex')['Survived'].sum()
```

```python
df.loc[(df['Survived']) & (df['Age']<18) & (df['Sex'])]
```

Спасли **61 пассажира**, чей возраст меньше **18 лет**

**Данные выжившего пассажира с максимальным возрастом**

```python
df.loc[(df['Survived']) & (df['Age']==df['Age'].max())]
```

```python
# Максимальный цена билета
print(df['Fare'].max())
# Минимальный цена билета
print(df['Fare'].min())
#Средняя цена билета
print(df['Fare'].mean())
```

```python
df[df['Fare'] ==0]
```

выше указаны данные пассажиров, которые не платили за билтеы, скорее всего это **экипаж Титаника**

```python
df[df['Fare'] ==(df['Fare'].max())]
```

Данные о пассажирах, которые заплатили самую высокую цену за билеты, при этом просмотрев стоблец **Survived** видно, что эти **пассажиры еще и выжили**!!!
